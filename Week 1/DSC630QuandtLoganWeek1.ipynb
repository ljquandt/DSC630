```python
#load in necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

The dataset I chose was an insurance dataset found on Kaggle (https://www.kaggle.com/datasets/mirichoi0218/insurance) that 7 columns and 1338 rows. The dataset contains information about individuals and factors that may influence their medical claim amounts. The columns are identified as the following:

age: age of primary beneficiary

sex: insurance contractor gender, female, male

bmi: Body mass index, providing an understanding of body, weights that are relatively high or low relative to height,
objective index of body weight (kg / m ^ 2) using the ratio of height to weight, ideally 18.5 to 24.9

children: Number of children covered by health insurance / Number of dependents

smoker: Smoking

region: the beneficiary's residential area in the US, northeast, southeast, southwest, northwest.

charges: Individual medical costs billed by health insurance

Questions:

1.) Do smokers have higher claims cost?

2.) Does age increase claims?

3.) Is region a factor in claims?

```python
#load in CSV file and ensure it came through correctly
df = pd.read_csv('insurance.csv')
df.head()
```

### Create a Histogram from your data

```python
#create histogram for smoking status and charges
sns.histplot(data=df[df['smoker'] == 'yes'], x='charges', label='Smoker', color = 'red')
sns.histplot(data = df[df['smoker'] == 'no'], x='charges', label='Non-Smoker', color='blue')
plt.title('Medical Cost Histogram - Smoking Status')
plt.xlabel('Claim Dollars')
plt.ylabel('Count')
plt.show()
```

### Create a boxplot from your data.

```python
#create boxplot of the medical charges
sns.boxplot(x=df['charges'])
plt.title("Medical Charges")
plt.show()
```

### Create a bivariate plot from your data

```python
#create bivariate scatter plot
sns.scatterplot(x=df['age'], y=df['charges'])
plt.title("BMI and Medical Charges")
plt.xlabel("BMI")
plt.ylabel('Charges')
plt.show()
```

### Create any additional visualizations that will help to answer the question(s) you want to answer

```python
#create barplot for region and charges
sns.barplot(x = df['region'], y= df['charges'])
plt.title('Region vs Charges')
plt.xlabel('Region')
plt.ylabel('Charges')
plt.show()
```

### Summarize your results and make a conclusion. Explain how you arrived at this conclusion and how your visualizations support your conclusion.

The Histogram clearly shows that smokers have higher claims than non-smokers. No non-smoker had claims above 40000 dollars while some smokers had claims in excess of 60000 dollars.

The boxplot clearly shows that median claims are around 10000 dollars with a few outliers exceeding 60000 dollars which per our histogram are likely smokers.

The bivariate scatterplot shows that claims have a linear relationship with age and tend to increase as age does. The data also appears to be clustered.

The barplot tells us that the southeast region has on average higher claims than the other regions. Northeast also has higher claims on average than the northwest or southwest which appear to be very close.

Overall, I was able to answer my questions showing that smokers do indeed have higher claims than non-smokers and age does lead to increased claims both of which are backed up by scientific data showing increased risk of chronic conditions. Lastly, region may have some impact on claims as the southeast has higher claims this may warrant further research into factors such as diet, activity levels and other factors that may lead to regional differences in health.
